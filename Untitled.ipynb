```python
#Removal of any Duplicate rows (if any)

counter = 0
r,c = original_df.shape

df1 = df.copy()
df1.drop_duplicates(inplace=True)
df1.reset_index(drop=True,inplace=True)

if df1.shape==(r,c):
    print('\n\033[1mInference:\033[0m The dataset doesn\'t have any duplicates')
else:
    print(f'\n\033[1mInference:\033[0m Number of duplicates dropped ---> {r-df1.shape[0]}')
```

```python
#Check for empty elements

nvc = pd.DataFrame(df1.isnull().sum().sort_values(), columns=['Total Null Values'])
nvc['Percentage'] = round(nvc['Total Null Values']/df1.shape[0],3)*100
print(nvc)
```

```python
nvc.head()
```

```python
# Converting categorical columns to numeric
import pandas as pd

ecc = nvc[nvc['Percentage'] != 0].index.values  # Exclude columns with certain condition
dcc = [i for i in df.columns if i not in ecc]   # Columns to keep
df3 = df1[dcc]                                  # New DataFrame with selected columns
fcc = [i for i in cf if i not in ecc]           # Filtered categorical features

# Flags to track first occurrence of encoding types
oh = True  # One-Hot encoding flag
dm = True  # Dummy encoding flag

# Transform categorical features
for i in fcc:
    # One-Hot Encoding for binary categorical features
    if df3[i].nunique() == 2:
        if oh:
            print("\033[1m\nOne-Hot Encoding on features:\033[0m")
        print(i)
        oh = False  # One-Hot Encoding flag set to False after first occurrence
        df3[i] = pd.get_dummies(df3[i], drop_first=True, prefix=str(i))  # Use loc to avoid warnings
    
    # Dummy Encoding for features with more than 2 but fewer than 17 unique values
    elif 2 < df3[i].nunique() < 17:
        if dm:
            print("\n\033[1mDummy Encoding on features:\033[0m")
        print(i)
        dm = False  # Dummy Encoding flag set to False after first occurrence
        df3 = pd.concat([df3.drop([i], axis=1), 
                         pd.get_dummies(df3[i], drop_first=True, prefix=str(i))], axis=1)  # Concatenate encoded columns

# Output final DataFrame shape
df3.shape
```

```python
# Removal of outliers using IQR
df4 = df3.copy()

for col in df4.select_dtypes(include=['float64', 'int64']).columns:  # Consider only numerical columns
    if df4[col].nunique() >= 12:  # Only apply IQR to features with sufficient unique values
        Q1 = df4[col].quantile(0.25)
        Q3 = df4[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter rows within bounds
        df4 = df4[(df4[col] >= lower_bound) & (df4[col] <= upper_bound)]

# Resetting index after outlier removal
df4 = df4.reset_index(drop=True)

# Displaying the first few rows of the dataset after outlier removal
display(df4.head())

# Printing before and after sample counts
print('\n\033[1mInference:\033[0m Before removal of outliers, the dataset had {} samples.'.format(df1.shape[0]))
print('\033[1mInference:\033[0m After removal of outliers, the dataset now has {} samples.'.format(df4.shape[0]))
```

```python
from imblearn.over_sampling import SMOTE

# Copy the dataset
df5 = df4.copy()

# Checking original class distribution
print('Original class distribution:')
print(df5[target].value_counts())

# Separating features and target
X = df5.drop([target], axis=1)
Y = df5[target]

# Applying SMOTE to balance the target classes
smote = SMOTE(random_state=42)
X_resampled, Y_resampled = smote.fit_resample(X, Y)

# Reconstruct the DataFrame with the resampled data
df5_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df5_resampled[target] = Y_resampled

# Displaying the class distribution after applying SMOTE
print('\nClass distribution after applying SMOTE Technique:')
print(df5_resampled[target].value_counts())
```

```python
#Final Dataset size after performing Preprocessing

df = df5.copy()
plt.title('Final Dataset Samples')
plt.pie([df.shape[0], original_df.shape[0]-df4.shape[0], df5.shape[0]-df4.shape[0]], radius = 1, shadow=True,
        labels=['Retained','Dropped','Augmented'], counterclock=False, autopct='%1.1f%%', pctdistance=0.9, explode=[0,0,0])
plt.pie([df.shape[0]], labels=['100%'], labeldistance=-0, radius=0.78, shadow=True, colors=['powderblue'])
plt.show()

print('\n\033[1mInference:\033[0mThe final dataset after cleanup has {} samples & {} columns.'.format(df.shape[0], df.shape[1]))
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into training & testing sets
df = df5.copy()

# Separating features and target
X = df.drop([target], axis=1)
Y = df[target]

# Train-test split
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=0)

# Output the shapes of the splits
print('Original set  ---> ', X.shape, Y.shape)
print('Training set  ---> ', Train_X.shape, Train_Y.shape)
print('Testing set   ---> ', Test_X.shape, Test_Y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Feature Scaling (Standardization)
std = StandardScaler()

# Standardization on Training set
print('\033[1mStandardization on Training set'.center(100))
Train_X_std = std.fit_transform(Train_X)
Train_X_std = pd.DataFrame(Train_X_std, columns=X.columns)
display(Train_X_std.describe())

# Standardization on Testing set
print('\n', '\033[1mStandardization on Testing set'.center(100))
Test_X_std = std.transform(Test_X)
Test_X_std = pd.DataFrame(Test_X_std, columns=X.columns)
display(Test_X_std.describe())
```
